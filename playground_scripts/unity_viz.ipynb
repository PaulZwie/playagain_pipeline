```python
# rms of the emg data stored in EMG_Ch0 ... EMG_Ch31
rms = np.sqrt(np.mean(df.iloc[:, 8:40]**2, axis=1))
```

```python
#plot the data with groundtruth as green background
plt.figure(figsize=(15, 5))
plt.plot(df["Timestamp"], rms, label="EMG RMS")
plt.fill_between(df["Timestamp"], 0, max(rms), where=df["GroundTruth"] == 1, color="green", alpha=0.3, label="Ground Truth")
```
