```python
# read in the csv files and find a threshold, for the emg data EMG_Ch0 to EMG_Ch31 that best separates the data into two classes below and above, below was the rest and above avtive only one threshold per file
thresholds = []
for file in csv_files:
    df = pd.read_csv(file)
    # find the mean of the emg data for each channel
    emg_data = df.iloc[:, 8:40]
    mean_emg = emg_data.mean()
    # find the threshold for each file
    threshold = mean_emg.mean() + 2 * mean_emg.std()
    thresholds.append(threshold)
```

```python
# the thresholds should be in mv instead of v, so convert them to mv
thresholds_mv = [threshold * 1000 for threshold in thresholds]
```
